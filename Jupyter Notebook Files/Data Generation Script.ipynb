```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pickle
import pandas as pd  

N_train = 5*1000 # Five thousand
N_test = 2*1000 # Two thousand
N_val = 2*1000 # Two thousand
```

# Dataset1:

Characteristics:
1. Single Variable Input
2. Homoscedastic Uncertainty
3. Symmetric Uncertainty
4. Low Uncertainty
5. Trigonometric Relationships between Input and Output

```python
Noise_Mag = 0.1
figure_size = (4, 4)

N = N_train
X1 = np.arange(0,2*math.pi,2*math.pi/N) # Range = 0 to 2*pi
y = np.sin(2*X1)+ Noise_Mag*(np.random.rand(N)-1/2)

plt.figure(figsize=figure_size) 
plt.title('Target vs X1')
plt.scatter(X1, y, color='k',s=.2)
plt.show()

Data = np.column_stack((X1,y))
print('Train Set', Data.shape)
dict = {'Input1': X1, 'Output': y}  
df = pd.DataFrame(dict) 
df.to_csv('Data1_Train.csv', header=True, index=False) 
output = open('Data1_Train.pkl', 'wb')
pickle.dump(Data, output)
output.close()


N = N_test
X1 = np.arange(0,2*math.pi,2*math.pi/N) # Range = 0 to 2*pi
y = np.sin(2*X1)+ Noise_Mag*(np.random.rand(N)-1/2)

plt.figure(figsize=figure_size) 
plt.title('Target vs X1')
plt.scatter(X1, y, color='k',s=.2)
plt.show()

Data = np.column_stack((X1,y))
print('Test Set',Data.shape)
dict = {'Input1': X1, 'Output': y}  
df = pd.DataFrame(dict) 
df.to_csv('Data1_Test.csv', header=True, index=False) 
output = open('Data1_Test.pkl', 'wb')
pickle.dump(Data, output)
output.close()

N = N_val
X1 = np.arange(0,2*math.pi,2*math.pi/N) # Range = 0 to 2*pi
y = np.sin(2*X1)+ Noise_Mag*(np.random.rand(N)-1/2)

plt.figure(figsize=figure_size) 
plt.title('Target vs X1')
plt.scatter(X1, y, color='k',s=.2)
plt.show()

Data = np.column_stack((X1,y))
print('Val Set',Data.shape)
dict = {'Input1': X1, 'Output': y}  
df = pd.DataFrame(dict) 
df.to_csv('Data1_Val.csv', header=True, index=False) 
output = open('Data1_Val.pkl', 'wb')
pickle.dump(Data, output)
output.close()
```

# Dataset4:
Characteristics:
1. Three-Variable Input
        a. First Variable determine a rough Point Prediction
        b. First Two Variables determine the magnitude of Uncertainty
        c. Input combinations are not related to each other
        d. Third Variable is not related to output
2. Heteroscedastic Uncertainty
3. Asymmetric Uncertainty (The density of uncertainty towards y-axis is not the same/symmetric)
4. Medium Uncertainty
5. Polynomial Relationships between Inputs and Output

```python
N = N_train
Noise_Mag = 3
X1 = np.random.rand(N)*10
X2 = np.random.rand(N)
X3 = np.random.rand(N)*2*math.pi
prediction = X1*(10-X1)
uncertainty = Noise_Mag*np.power(np.random.rand(N),2)*X2*(1-X2)*X1*(10-X1)
y = prediction + uncertainty

plt.figure(figsize=figure_size) 
plt.scatter(X1, prediction, color='k',s=.2)
plt.title('Rough Prediction vs X1')
plt.show()

plt.figure(figsize=figure_size) 
plt.title('Uncertainty vs X2')
plt.scatter(X2, uncertainty, color='k',s=.2)
plt.show()

plt.figure(figsize=figure_size) 
plt.title('Target vs X1')
plt.scatter(X1, y, color='k',s=.2)
plt.show()

plt.figure(figsize=figure_size) 
plt.title('Target vs X2')
plt.scatter(X2, y, color='k',s=.2)
plt.show()

plt.figure(figsize=figure_size) 
plt.title('Target vs X2*(1-X2)*X1*(10-X1)')
plt.scatter(X2*(1-X2)*X1*(10-X1), y, color='k',s=.2)
plt.show()

Data = np.column_stack((X1,X2, X3,y))
print('Train Set',Data.shape)
dict = {'Input1': X1, 'Input2': X2, 'Input3': X3, 'Output': y}  
df = pd.DataFrame(dict) 
df.to_csv('Data4_Train.csv', header=True, index=False) 
output = open('Data4_Train.pkl', 'wb')
pickle.dump(Data, output)
output.close()


N = N_test
X1 = np.random.rand(N)*10
X2 = np.random.rand(N)
X3 = np.random.rand(N)*2*math.pi
prediction = X1*(10-X1)
uncertainty = Noise_Mag*np.power(np.random.rand(N),2)*X2*(1-X2)*X1*(10-X1)
y = prediction + uncertainty

Data = np.column_stack((X1,X2, X3,y))
print('Test Set',Data.shape)
dict = {'Input1': X1, 'Input2': X2, 'Input3': X3, 'Output': y}  
df = pd.DataFrame(dict) 
df.to_csv('Data4_Test.csv', header=True, index=False) 
output = open('Data4_Test.pkl', 'wb')
pickle.dump(Data, output)
output.close()

N = N_val
X1 = np.random.rand(N)*10
X2 = np.random.rand(N)
X3 = np.random.rand(N)*2*math.pi
prediction = X1*(10-X1)
uncertainty = Noise_Mag*np.power(np.random.rand(N),2)*X2*(1-X2)*X1*(10-X1)
y = prediction + uncertainty

Data = np.column_stack((X1,X2, X3,y))
print('Val Set',Data.shape)
dict = {'Input1': X1, 'Input2': X2, 'Input3': X3, 'Output': y}  
df = pd.DataFrame(dict) 
df.to_csv('Data4_Val.csv', header=True, index=False) 
output = open('Data4_Val.pkl', 'wb')
pickle.dump(Data, output)
output.close()
```
